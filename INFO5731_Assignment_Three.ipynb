```python
import pandas as pd
from nltk.corpus import stopwords
# Importing the file from Assignment 2 and saving it in a dataframe
df_abstract = pd.read_csv('romal_inclass4.csv', encoding= 'unicode_escape')
df_abstract
```

```python
# Before beginning to answer question 1, we need to clean the Abstract column
import nltk
nltk.download('stopwords')
from textblob import TextBlob

# Lower casing
df_abstract['Abstract'] = df_abstract['Abstract'].astype(str)
df_abstract['Abstract'] = df_abstract['Abstract'].apply(lambda x: " ".join(x.lower() for x in x.split()))

# Removing Punctuation
df_abstract['Abstract'] = df_abstract['Abstract'].str.replace('[^\w\s]','')
df_abstract.head()


```

```python
# Removing unnecessary words
stop = stopwords.words('english')
df_abstract['Abstract'] = df_abstract['Abstract'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df_abstract.head()
```

```python
import nltk
nltk.download('punkt')
# Common word removal
freq = pd.Series(' '.join(df_abstract['Abstract']).split()).value_counts()[:10]
# freq
freq = list(freq.index)
df_abstract['Abstract'] = df_abstract['Abstract'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
df_abstract.head()
# Rare words removal
freq = pd.Series(' '.join(df_abstract['Abstract']).split()).value_counts()[-10:]
# freq
freq = list(freq.index)
df_abstract['Abstract'] = df_abstract['Abstract'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
# # Spelling Correction
# df_abstract['Abstract'].apply(lambda x: str(TextBlob(x).correct()))

# Tokenization
df_abstract['Abstract'] = df_abstract['Abstract'].apply(lambda x: TextBlob(x).words)
df_abstract.head()
df_abstract.to_csv('romal_cleanfile.csv')
df_abstract.head()
```

```python
# Frequency of n-grams

# Using the itertools library instead of the for-loop for easy use
import itertools
ngrams_words = []

for sentence in df_abstract['Abstract']:
  ngrams_words.append(sentence)   

words_1 = [x for x in ngrams_words if x != []]
iterations = list(itertools.chain.from_iterable(words_1))
trigrams = nltk.trigrams(iterations)
frequency_dist = nltk.FreqDist(trigrams)
trigrams_dict = dict(frequency_dist)
for token_words in trigrams_dict:

  # Printing the dictionary
    print( str(token_words) + ': '+ str(frequency_dist[token_words]))
```
