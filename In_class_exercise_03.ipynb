```python
# You code here (Please add comments in the code):
# The dataset used for this exercise is extracted from the kaggle website
import pandas as pd
import numpy as np
import spacy
```

```python
# Understanding the dataset
tweet=pd.read_csv('corona_tweet.csv')
tweet
```

```python
# Understandig the dataset
# We can see from the summary below that the negative tweets were tweeted maximum during COVID times which is similar to what
# one would also predict.
tweet.sentiment.value_counts()
```

```python
# To convert the sentiments into numbers for easy analysis, I have set 
#'Extremely Positive' = 2
#'Extremely Negative' = 0
#'Positive' = 2
#'Negative' = 0
#for all the other sentiments = 0
def numbers(sentiment):
    if sentiment == "Extremely Positive":
        return 2
    elif sentiment == "Extremely Negative":
        return 0
    elif sentiment == "Positive":
        return 2
    elif sentiment == "Negative":
        return 0
    else:
        return 1
# Now applying the above function to our dataframe
tweet.sentiment = tweet.sentiment.apply(lambda x : numbers(x))
tweet
# The dataframe below this has sentiments converted to numerical form
```

```python
# 1. Lower case
tweet['Originaltweet'] = tweet['Originaltweet'].apply(lambda x: " ".join(x.lower() for x in x.split()))
tweet['Originaltweet'].head()
```

```python
#2. Special characters count
import re
def special_char(data, remove_digits=True):
    reg_ex = r'[^a-zA-z0-9\s]'
    data = re.sub(reg_ex,'',data)
    return data
tweet['Originaltweet'] = tweet['Originaltweet'].apply(special_char)
tweet
```

```python
# 3. Count the numbers in the dataset
tweet['numerics'] = tweet['Originaltweet'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
tweet
```

```python
# Count of stopwords
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop = stopwords.words('english')
tweet['stopwords'] = tweet['Originaltweet'].apply(lambda x: len([x for x in x.split() if x in stop]))
tweet
```

```python
# 4. Removing stop words
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
tweet['Originaltweet'] = tweet['Originaltweet'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
tweet
```
